```python
import random
dic = {2: 2.78, 3:5.56, 4:8.33,5:11.11,6:13.89,7:16.67,8:13.89,9:11.11,10:8.33,11:5.56,12:2.78}
print("\nDictionary with the use of Integer Keys: ")
print(dic)

```

```python
from itertools import product

two_dice_sums = [a + b for a, b in product(range(1, 7), repeat=2)]

def dice(n):
    return [random.choice(two_dice_sums) for _ in range(n)]
print(*dice(1000), sep='\n')
#rolls=dice(1000)
```
